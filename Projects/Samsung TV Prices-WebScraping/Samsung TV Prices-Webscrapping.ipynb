### Webscraping of SAMSUNG TV prices and its rating  

* This is a technique to retrieve large amounts of data from the internet. 
* This technique is highly useful in competitive pricing.there are different techniques involved in scrapping but the most common one is beautifulsoup. 
* The data stored are unstructured and it is then converted to structured format before EDA

#### Importing libraries required  for webscraping 

```python
import requests 
from bs4 import BeautifulSoup
```

```python
import numpy as np 
import pandas as pd 
```

```python
url="https://www.flipkart.com/search?q=samsung+tv"
```

##### Checking whether scrapping is allowed 

```python
r=requests.get(url)
r.status_code
```

#### 200 indicates link flipkart allows scrapping

```python
print(r.text)
```

```python
soup=BeautifulSoup(r.content,"lxml")
```

```python
soup
```

```python
print(soup.prettify())
```

#### Find the title 

```python
soup.title
```

```python
soup.title.string
```

#### Find the divisions 

```python
soup.div
```

```python
soup.a
```

```python
soup.find_all("a")
```

```python
all_link=soup.find_all("a")
for r in all_link:
    print(r.get("href"))
```

```python
al_tables=soup.find_all("table")
print(al_tables)
```

```python
print(soup.prettify())
```

#### scraping all products 

```python
model=soup.findAll("div",{'class':'_4rR01T'})
```

```python
model
```

```python
productname=[]
for i in model:
    p=i.text
    productname.append(p)
print(productname)
```

### Scrapping prices of all products

```python
prices=soup.findAll('div',{'class':"_30jeq3 _1_WHN1"})
```

```python
print(prices)
```

#### scrapping the rating of products 

```python
r=soup.findAll('div',class_="_3LWZlK")
```

```python
rating=[]
for i in r:
    ra=i.text
    rating.append(ra)
print(rating)
```

#### So the data set of 24 rows is ready 

```python
d={"productname":productname,"price":prices,"rating":rating}
tv_data=pd.DataFrame(d)
tv_data
```
