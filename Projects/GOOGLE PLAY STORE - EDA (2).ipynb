##### Free installation was much more than the paid and to confirm visualisation is done 

```python
plt.figure(figsize=(10,10))
labels=GPS_sample['Type of access'].value_counts(sort=True).index
sizes=GPS_sample['Type of access'].value_counts(sort=True)
explode=(0.5,0)
colors=['green','blue']
plt.pie(sizes,labels=labels,explode=explode,colors=colors,shadow=True,startangle=0)
plt.title('Installation v/s Type of access',size=20)
plt.legend()
plt.show()
```
