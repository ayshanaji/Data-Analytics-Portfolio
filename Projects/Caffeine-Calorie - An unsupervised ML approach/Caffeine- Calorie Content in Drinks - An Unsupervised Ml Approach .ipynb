### Part 7: <u>Creating a model <u>
* Model can be created to divide the drinks on the basis of Caffeine and Calories

```python
plt.scatter(df_sort.Caffeine_100ml,df_sort.Calories_100ml)
```

* The data looks clustered not in a uniform form 
    

### Un Supervised ML Approach- Kmeans


* Unsupervised machine learning is the process of inferring underlying hidden patterns from historical data. Within such an approach, a machine learning model tries to find any similarities, differences, patterns, and structure in data by itself. No prior human intervention is needed.

```python
from sklearn.cluster import KMeans
```

```python
k=KMeans(n_clusters=4)
```

```python
x=df_sort.drop(['type'],axis=1)
```

```python
x
```

```python
sse=[]
kr=range(1,10)
for k in kr:
    km = KMeans(n_clusters = k)
    km.fit(x)
    sse.append(km.inertia_)
```

```python
sse
```

#### Plotting the elbow:
Within-Cluster-Sum-of-Squares (WCSS) is the sum of squared distance between each point and the centroid in a cluster. When we plot the WCSS with the K value, the plot looks like an Elbow. As the number of clusters increases, the WCSS value will start to decrease.

```python
plt.xlabel('K')
plt.ylabel("Sum of Squares Errors")
plt.plot(kr,sse)
plt.show()
```

```python
y=k.fit_predict(x)
y
```

```python
k.cluster_centers_
```

```python
df_sort['cluster']=y
df_sort
```

```python
sf0=df_sort[df_sort.cluster==0]
sf1=df_sort[df_sort.cluster==1]
sf2=df_sort[df_sort.cluster==2]
sf3=df_sort[df_sort.cluster==3]

```

```python
plt.scatter(sf0.Caffeine_100ml,sf0.Calories_100ml,color='red',label='cluster1')
plt.scatter(sf1.Caffeine_100ml,sf1.Calories_100ml,color='green',label='cluster2')
plt.scatter(sf2.Caffeine_100ml,sf2.Calories_100ml,color='blue',label='cluster3')
plt.scatter(sf3.Caffeine_100ml,sf3.Calories_100ml,color='cyan',label='cluster4')
plt.scatter(k.cluster_centers_[:, 0], k.cluster_centers_[:, 1], s = 400, color = 'black', marker="*",label = 'Centroids')

plt.title('Clusters of Drinks')
plt.xlabel('Caffeine Content')
plt.ylabel('Calories Present')
plt.legend()
plt.show()
```
