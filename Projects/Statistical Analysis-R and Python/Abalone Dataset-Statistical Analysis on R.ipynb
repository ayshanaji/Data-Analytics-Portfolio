### Measures of central tendency

* Central tendency tells you about the centers of the data. Useful measures include the mean, median, and mode.

```r
print("Average Rings for Abalone: ")
mean(d$Rings)
print("Median Rings for Abalone: ")
median(d$Rings)
print("Modal Rings for Abalone: ")
mode(d$Rings)
```

#### A quantile is a statistical term describing the divisions of observations into four defined intervals. Divides data into three parts - lower quantile(25%), median(50%) and upper quantile (75%). Graphically it can be easily visualized through Boxplot.

```r
boxplot(d$Rings)
```

```r
quantile(d$Rings,c(0.25,.50,.75))
```

#### Measures of dispersion

* Variability tells you about the spread of the data. Useful measures include variance and standard deviation.

```r

print("The variance of Rings is given by:")
var(d$Rings)
print("The Standard deviation of Rings is given by:")
sd(d$Rings)
```

```r
hist.default(d$Rings,col="cyan")
```

### Corelation and Covariance

* Correlation or joint variability tells you about the relation between a pair of variables in a dataset. 
* Useful measures include covariance and the correlation coefficient.
* Corelation measures association.
* Positive correlation indicates the variables are in same direction otherwise opposite direction and zero indicates no relation.

```r
print("The correlation between Rings and shellweight:")
cor(d$Rings,d$Shell.weight)
```

#### It can be easily visualised by a scatterplot

```r
plot(d$Rings~d$Shell.weight,col='blue')
```

##### Covariance measures direct relationship between variables.

```r
print("The covariance between the rings and shell weight:")
cov(d$Rings,d$Shell.weight)
```

### Statistical tests
* There are many tests under statistics and t test is a simple and commony used test to predict the average using hypothesis rule

```r
t.test(d$Rings)
```

#### In this T-test, the P-value is <2.2e-16 and the confidence interval is 95%. It also shows the mean value of 9.933684.
##### This T-test shows the Alternate hypothesis is true in the hypothesis testing process.
