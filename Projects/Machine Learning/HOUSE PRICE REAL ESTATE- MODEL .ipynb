#### Now filling rest of the NaN values with the possible values 

```python
med_bath=df['bath'].median()
med_bath
```

```python
med_balcony=df['balcony'].median()
med_balcony
```

```python
mean_size=df['BHK'].mean()
mean_size
```

```python
df['bath']=df['bath'].fillna(med_bath)
df['balcony']=df['balcony'].fillna(med_balcony)
df['BHK']=df['BHK'].fillna(mean_size)
```

```python
df.BHK=df.BHK.astype(int)
```

```python
df.bath=df.bath.astype(int)
```

```python
df.balcony=df.balcony.astype(int)
```

```python
def squareft(ft):
    if ft<2000:
        return 1
    elif ft<10000:
        return 2
    elif ft<50000:
        return 3
    else:
        return 4
```

```python
df['totalsqft']=df['totalsqft'].apply(squareft)
```

#### From the scatterplot 
* with specific area,size of BHK,bath,balcony and the corresponding price varies 
* So total price gets affected.
* so data can be figured out using Linear Multiple Regression 

```python
from sklearn import linear_model 
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['areatype','BHK','totalsqft','bath','balcony']],df.totalprice)
```

```python
reg.predict([[2,3,2300,2,1]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

#### To check whether the value obtained is correct , manual check can be done using the formula
*  y=m1x1 + m2x2 +m3x3 + b 
where m1,m2,m3 are coeficients or slope of independent variables
        b is the intercept 
* for that reg.coef_ and reg.intercept_ can be used
