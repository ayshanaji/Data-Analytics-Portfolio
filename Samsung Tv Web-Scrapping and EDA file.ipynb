#### This is a technique to retrieve large amounts of data from the internet. This technique is highly useful in competitive pricing.there are different techniques involved in scrapping but the most common one is beautifulsoup. The data stored are unstructured and it is then converted to structured format before EDA

#### importing libraries required only for webscraping is beautifulsoup and the other libraries required for EDA will be imported later 

```python
import requests 
from bs4 import BeautifulSoup
```

#### Copy the url 

```python
url="https://www.flipkart.com/search?q=samsung+tv&as=on&as-show=on&otracker=AS_Query_PredictiveAutoSuggest_9_0_na_na_na&otracker1=AS_Query_PredictiveAutoSuggest_9_0_na_na_na&as-pos=9&as-type=PREDICTIVE&suggestionId=samsung+tv&requestId=9c7b7c3d-ab49-403e-a38a-9420a874b83c"
```

##### Checking whether scrapping is allowed 

```python
r=requests.get(url)
r.status_code
```

#### 200 indicates link flipkart allows scrapping

```python
print(r.text)
```

```python
soup=BeautifulSoup(r.content,"lxml")
```

#### scraping all products 

```python
model=soup.findAll("div",{'class':'_4rR01T'})
```

```python
model
```

```python
productname=[]
for i in model:
    p=i.text
    productname.append(p)
print(productname)
```

### scrapping prices of all products

```python
prices=soup.findAll('div',{'class':"_30jeq3 _1_WHN1"})
```

```python
print(prices)
```

#### scrapping the rating of products 

```python
r=soup.findAll('div',class_="_3LWZlK")
```

```python
rating=[]
for i in r:
    ra=i.text
    rating.append(ra)
print(rating)
```
